# READING A VIDEO STREAM FROM CAMERA (FRAME BY FRAME)

video is a series of frame and each frame is image 

```python
import cv2 
```

```python
cap = cv2.VideoCapture(0)
# here we pass the id and by default 0 is our webcam id , in case of multiple webcam -> it depends
```

```python
while True:
    ret, frame = cap.read()
    # here read returs tow things boolean and frame -> if ret is false then frame is not captured properly 
    # due to multiple reason -> cam is not started and get delayed to show up.
    if ret == False :
        continue
    cv2.imshow('Video Frame',frame)
    
    
    #=====================
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow('GRAY Video Frame',gray_frame)
    
    #================
    
    
    
    # wait for user input -q then you will stop the looop
    
    key_pressed = cv2.waitKey(1) & 0xFF
    # here waitKey returns 32 bit and 0xFF is a 8 bit integer with all 1's and when we do and with it 
    # -> we get the 8 bit integer then we check with 8 bit of ord ('q') -> to terminate the code.
    if key_pressed == ord('q'):
        break;
         # it will stop working when only q key is pressed
    # ord('q') return the ASCII value of input character.
cap.release() # it will release the device
cv2.destroyAllWindows() # it will destroy all the windows
```

# FACE DETETCTION USING OpenV
here we will use 2 things

pre-trained Model -> HaarCascades (CNN Model) (alraedy trained for face detetction)

now we will use
KNN for classifying the face 



here we will use already trained model to figure out wether the given object is a face or not
and we will use KNN to classify this 


HOW HAAR_CASCADE works 
it basically detects some features like -> we have eyes or not , do we have nose or not
so on  and keep detecting -> which feature is available or not. 


it is an old method now moder method is CNN.

```python
import cv2 
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
    # here we are creating an object of CascadeClassiier and training it on facial data using above xml
    if ret == False :
        continue
        
    #now we will provide this frames to our model in order to detect whether the given object is face or not
    
    faces = face_cascade.detectMultiScale(gray_frame , 1.3 , 5)# using gray image only for training
    
    # it takes 3 parameters image, scaling factor, NoofNeigbour
    
    """
        what are these parameters => ScalingFactor specifies how much image size is reduced at each image scale i.e
        at 1.3 -> image size is reduced by 30% and in case 1.05 it is reduced by 5%
        why we want scalling -> in order to train our images keeping same size on which our model is trained.
        
        Neighbour is used to retain data of multiple neighbour rectange to detect feature 
        high value will give high quality but less detection
         3-6 are good values
    """
    
    # ABOVE METHOD -> will return a x,y coordinate and widht and height of face if found and if there are multiple faces then
    # it will give an list of all four values
    
    for (x,y,w,h) in faces :
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
        # images , starting coordinate and ending coordinate , color and widht
    cv2.imshow('Video Frame',frame)
   #cv2.imshow('GRAY Video Frame',gray_frame)
    
    key_pressed = cv2.waitKey(1) & 0xFF
    if key_pressed == ord('q'):
        break;
cap.release() 
cv2.destroyAllWindows() 
```

# FACE RECOGNITION Using OpenCV
Now we will try to recognize the Face 


Lets first store some face data

```python
import cv2
import numpy as np

#initialize the camera
cap = cv2.VideoCapture(0)

#face detection

face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")

skip = 0
face_data = []

while True:
    ret, frame = cap.read()
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
    
    if ret == False :
        continue
        
    
    faces = face_cascade.detectMultiScale(gray_frame , 1.3 , 5)
    faces  = sorted(faces , key = lambda f: f[2]*f[3])
    # here we are sorting multiple faces on the basis of area acquired by an image
    
    # larger image comes first
    for face in faces[-1:] :
        x,y,w,h = face
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
        #extract the image 
        offset = 10
        face_section = frame[y-offset : y+offset + h , x-offset : x+offset + w  ]
        face_section = cv2.resize(face_section,(100,100) )
        # resizing into fized size in order to keep features constant
        skip+=1
        if skip %10 ==0:
            face_data.append(face_section)
            print(len(face_data))
        
    cv2.imshow('Video Frame',frame)
    
    key_pressed = cv2.waitKey(1) & 0xFF
    if key_pressed == ord('q'):
        break;
        
# convert  our face list into numpy array
face_data = np.asarray(face_data)
face_data = face_data.reshape((face_data.shape[0],-1))

# saving the data into the file system
#np.save("gaurang.npy",face_data)
#np.save("gaurang1.npy",face_data)
np.save("gaurang2.npy",face_data)
cap.release() 
cv2.destroyAllWindows() 
```

SO FAR WE HAVE COLLECTED OUR FACE DATA FOR TRAINING THE MODEL -> NOW WE WILL TRAIN OUR MODEL FOR FACE RECOGNITION.
WE CAN USE VARIOUS MODEL LIKE LINEAR REGRESSION , LOGISTIC , SVM , KNN etc.


NOW WE WILL USE THE SAVED DATA TO TRAIN OUR MODEL AND THEN USE KNN TO CLASSIFY THE FACE TO WHICH IT RESEMBLE THE MOST.


```python
import cv2
import numpy as np
import os


#initialize the camera
cap = cv2.VideoCapture(0)

#face detection

face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")

skip = 0
face_data = []

labels = []

class_id = 0 # labels for given file
name = {} # mapping between id and name


# data preparation :- getting stored data

for fx in os.listdir("./"):
    if fx.endswith(".npy"):
        # mapping name with class id 
        name[class_id] = fx[:-4]
        data_item = np.load(fx)
        print(fx)
        face_data.append(data_item)

        # creatiing labels for the class
        target = class_id * np.ones((data_item.shape[0],))
        # basically one file can have mulitple face data of one person
        #-> so we are mapping them to particular number for training purpose

        class_id += 1
        labels.append(target)
face_dataset = np.concatenate(face_data,axis=0)
face_labels = np.concatenate(labels,axis=0)

print(face_dataset.shape)
print(face_labels.shape)

```

```python
from sklearn.neighbors import KNeighborsClassifier
# now training our captured data
```

```python
knn = KNeighborsClassifier()
knn.fit(face_dataset,face_labels)
```

```python
# Testing

import cv2 
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
    if ret == False :
        continue
    
    faces = face_cascade.detectMultiScale(gray_frame , 1.3 , 5)
    
    for face in faces[-1:] :
        x,y,w,h = face
        # getting face for prediction
        offset = 10
        face_section = frame[y-offset : y+offset + h , x-offset : x+offset + w  ]
        face_section = cv2.resize(face_section,(100,100) )
        # predicting the face
        label = knn.predict(face_section.reshape(1, -1))
        
        #display the screen and name around the face
        cv2.putText(frame , name[label[0]],(x,y-10),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2,cv2.LINE_AA)
        #refer documentation
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
        
    cv2.imshow("Faces",frame)
    key_pressed = cv2.waitKey(1) & 0xFF
    if key_pressed == ord('q'):
        break;
cap.release() 
cv2.destroyAllWindows() 
```
